```python
## open the file containing ToposText annotation in the NH
ToposText_NH = pd.read_csv("/Users/u0154817/OneDrive - KU Leuven/Documents/KU Leuven/PhD project 'Greek Spaces in Roman Times'/Data_Extraction/Outputs/topostext_places_NH.csv", delimiter=",")
```

```python
for i1,topostext in enumerate(ToposText_NH['tagged_entity']):
    
    print(i1,topostext)
    
    reference_topostext = ToposText_NH['reference'][i1]
    startpos_topostext = ToposText_NH['start_pos'][i1]
    topostext_id = ToposText_NH['topostext_id'][i1]
        
    for i2,flairNER in enumerate(flairNER_NH['token']):
        
        if flairNER_NH['reference'][i2] == reference_topostext:
            if flairNER_NH['start_pos'][i2] == startpos_topostext:
                
                if 'LOC' in flairNER_NH['flair_ner'][i2]: ## the entity was annotated by flairNER
                    flairNER_NH['topostext_id'][i2] = topostext_id ## add the ToposText ID
                    
                    print(i2, flairNER_NH['reference'][i2], flairNER_NH['start_pos'][i2], flairNER_NH['flair_ner'][i2], flairNER_NH['topostext_id'][i2])
                    
                else : 
                    
                    topostext = str(topostext)
                    topostext = topostext.split() ## split the ToposText annotation into tokens
                    
                    if len(topostext) == 1: ## it consists of a single token
                        
                        flairNER_NH['flair_ner'][i2] = 'B-LOC'
                        flairNER_NH['topostext_id'][i2] = topostext_id
                        
                        print(i2, flairNER_NH['reference'][i2], flairNER_NH['start_pos'][i2], flairNER_NH['flair_ner'][i2], flairNER_NH['topostext_id'][i2])
                    
                    else :
                        
                        flag = False
                        
                        for i3, token in enumerate(topostext):
                            
                            if i3 > 0:
                                
                                if flairNER_NH['flair_ner'][i2+i3] == 'B-LOC':
                                    
                                    flag = True ## the entity is annotated inside
                                    flairNER_NH['topostext_id'][i2+i3] = topostext_id
                                    
                                    print(i2+i3, flairNER_NH['reference'][i2+i3], flairNER_NH['start_pos'][i2+i3], flairNER_NH['flair_ner'][i2+i3], flairNER_NH['topostext_id'][i2+i3])
                                    
                        if flag == False:
                            
                            flairNER_NH['flair_ner'][i2] = 'B-LOC'
                            flairNER_NH['topostext_id'][i2] = topostext_id
                            
                            print(i2, flairNER_NH['reference'][i2], flairNER_NH['start_pos'][i2], flairNER_NH['flair_ner'][i2], flairNER_NH['topostext_id'][i2])
```
