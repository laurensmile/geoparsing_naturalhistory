```python
for i1, topostext_annotation in enumerate(ToposText_Places_NH['tagged_entity']): ## for each ToposText annotation
    print(i1,topostext_annotation)
        
    reference = ToposText_Places_NH['reference'][i1] ## get the reference (book, chapter, paragraph)
    start_position = ToposText_Places_NH['start_pos'][i1] ## get the start position
    
    for i2, flair_annotation in enumerate(Flair_NER_NH['Token']): ## for each token in the Gold Standard
        
        if Flair_NER_NH['Reference'][i2] == reference: ## if the reference is the same
            if Flair_NER_NH['Start_pos'][i2] == start_position: ## if the start position is the same
                                
                Flair_NER_NH['ToposText'][i2] = 'B-LOC' ## update the ToposText column
                Flair_NER_NH['ToposText_ID'][i2] = 'B-'+str(ToposText_Places_NH['ToposText_ID'][i1]) ## update the ToposText ID column
                
                topostext_annotation = topostext_annotation.split() ## split the text annotated in ToposText
                print(i1, topostext_annotation)
                if len(topostext_annotation) > 1: ## if the annotation contains more than one word
                                        
                    for i3, word in enumerate(topostext_annotation): ## for each word
                        
                        if i3 > 0: ## except the first one
                                                        
                            Flair_NER_NH['ToposText'][i2+i3] = 'I-LOC' ## update the corresponding ToposText column
                            Flair_NER_NH['ToposText_ID'][i2+i3] = 'I-'+str(ToposText_Places_NH['ToposText_ID'][i1]) ## update the corresponding ToposText ID column
```
